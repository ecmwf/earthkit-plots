## Cities in Japan

This example demonstrates how to add ancillary layers, including cities, to a map.

```python
import earthkit.plots as ekp

chart = ekp.Map(domain="Japan")

chart.land()
chart.ocean()
chart.urban_areas()

chart.borders()
chart.administrative_areas()
chart.gridlines()
chart.cities(adjust_labels=True)

chart.title("Cities in {domain}")

chart.show()
```
