```python
import earthkit.data

from earthkit.plots.interactive import Chart
```

```python
data = earthkit.data.from_source("sample", "timeseries.json")
```

```python
chart = Chart()
chart.box(data)
chart.title("ECMWF ensemble meteogram")
chart.line(data, aggregation="mean", line_color="grey")

# chart.show()  # Run this in an interactive session!
```
