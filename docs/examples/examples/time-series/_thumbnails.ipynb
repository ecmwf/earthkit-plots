```python
from pathlib import Path

import earthkit.data as ekd

import earthkit.plots as ekp

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")
```

```python
dataset = "reanalysis-era5-single-levels-timeseries"
request = {
    "variable": ["2m_temperature"],
    "location": {"longitude": -1, "latitude": 51.5},
    "date": ["1940-01-01/2025-12-01"],
    "data_format": "netcdf",
}

data = ekd.from_source("cds", dataset, request)
ds = data.to_xarray()
ds
```

```python
import earthkit.transforms as ekt
```

```python
yearly = ekt.resample(ds, frequency="Y")
```

```python
import matplotlib.pyplot as plt

chart = ekp.TimeSeries(figsize=(2, 2))
chart.line(yearly, units="celsius", color="#F94144", linewidth=1, drawstyle="spline")
chart.ax.set_frame_on(False)
plt.axis("off")
ticks = chart.xticks(
    frequency="20Y",
    format="%Y",
)

for tick in chart.ax.get_xticklabels():
    tick.set_fontsize(8)
for tick in chart.ax.get_yticklabels():
    tick.set_fontsize(8)

chart.save(THUMBNAILS_DIR / "timeseries-introduction.png", transparent=True)
```
