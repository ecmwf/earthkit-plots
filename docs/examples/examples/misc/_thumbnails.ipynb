```python
from pathlib import Path

import cartopy.crs as ccrs
import earthkit.data as ekd

import earthkit.plots as ekp

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")
```

```python
data = ekd.from_source(
    "mars",
    **{
        "channel": "3/2/1",
        "class": "od",
        "date": -1,
        "ident": "71",
        "instrument": "210",
        "param": "260512",
        "step": "12",
        "stream": "oper",
        "time": "00:00:00",
        "type": "ssd",
        "grid": "0.1/0.1",
    },
).to_fieldlist()
```

```python
chart = ekp.Map(
    crs=ccrs.NearsidePerspective(central_longitude=10, central_latitude=10),
    figsize=(2, 2),
)
chart.rgb_composite(data)
chart.save(THUMBNAILS_DIR / "rgb-composite.png", transparent=True)
```

```python
nuts_mean_2t = ekd.from_source("sample", "era5-nuts-mean-2t.geojson").to_pandas()
colors = [
    "#FFF0F0",
    "#FFD0CF",
    "#FFA8A6",
    "#FF7B78",
    "#F94144",
    "#E02E31",
    "#C41E21",
    "#A31014",
    "#7E060A",
    "#580006",
][::-1]
```

```python
style = ekp.styles.Style(
    levels=range(-10, 11),
    colors="Spectral_r",
    units="celsius",
)

chart = ekp.Map(domain=["France", "Spain"], figsize=(2, 2))

chart.choropleth(
    nuts_mean_2t,
    z="2t",
    metadata={"units": "K", "long_name": "2m temperature"},
    style=style,
)

chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR / "choropleth.png", transparent=True)
```
