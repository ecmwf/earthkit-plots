## Isolines - more complex styles

In this section, we will explore how to create reusable `Style objects`, allowing you to separate styling logic from the rest of your visualisation workflow. Reusable styles make your plots more consistent, maintainable, and adaptable to different datasets.

First, we will load the sample mean sea level pressure data from the ERA5 dataset, continuing from the previous example.

```python
import earthkit as ek

data = ek.data.from_source("sample", "era5-2t-msl-1985122512.grib")
pressure = data.sel(short_name="msl")
```

Now let's define our style. There is a special `earthkit.plots.styles.Contour` class which takes all of the `contour` parameters you will be familiar with if you have worked through the other examples on isolines. The only new thing we are introducing here is passing a matplotlib named colour map as the `linecolors` argument, which will result in our contours having different colours based on their level. You can also pass a list of colours or a single colour.

```python
style = ek.plots.styles.Contour(
    units="hPa",
    levels={"step": 4, "reference": 1000},
    linewidths=1,
    labels=True,
    linecolors="turbo",
)
```

Finally, we can use this style by simply passing it as the `style` argument to `Map.contour()`.

```python
chart = ek.plots.Map(domain="Europe")

chart.contour(pressure, style=style)
chart.legend()

chart.land()
chart.borders()
chart.coastlines()

chart.title()

chart.show()
```
