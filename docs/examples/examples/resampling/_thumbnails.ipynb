```python
from pathlib import Path

import earthkit.data as ekd
import numpy as np
from earthkit.geo import regrid

from earthkit.plots.resample import Bilinear, NearestNeighbour, Regrid

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")

data = ekd.from_source("sample", "era5-2t-msl-1985122512.grib").to_fieldlist()
pressure = data.sel({"parameter.variable": "msl"})
temperature = data.sel({"parameter.variable": "2t"})
```

```python
rgg = regrid(temperature, grid={"grid": "O120"})
```

```python
t2 = regrid(temperature, grid={"grid": [0.5, 0.5]})
```

```python
t4 = regrid(temperature, grid={"grid": [1, 1]})
```

```python
import earthkit.plots

# Example data (unstructured points)
points = np.random.rand(60, 2) * 10  # 500 random points
values = np.sin(points[:, 0]) + np.cos(points[:, 1])  # Some sample data

x = points[:, 0]
y = points[:, 1]
```

```python
style = earthkit.plots.styles.Style(
    levels=np.arange(-2, 2.1, 0.25),
    colors="Reds",
)
```

```python
chart = earthkit.plots.Map(figsize=(2, 2))

chart.point_cloud(x=x, y=y, z=values)
chart.pcolormesh(x=x, y=y, z=values, style=style, resample=Bilinear())
chart.scatter(x=x, marker="X", y=y, edgecolors="black", z=values, c=values, style=style)
chart.ax.set_frame_on(False)
chart.ax.set_extent([2, 8, 2, 8])
chart.save(THUMBNAILS_DIR / "resampling-bilinear.png", transparent=True)
```

```python
chart = earthkit.plots.Map(figsize=(2, 2))

chart.point_cloud(x=x, y=y, z=values)
chart.pcolormesh(x=x, y=y, z=values, style=style, resample=NearestNeighbour())
chart.scatter(x=x, marker="X", y=y, edgecolors="black", z=values, c=values, style=style)
chart.ax.set_frame_on(False)
chart.ax.set_extent([2, 8, 2, 8])
chart.save(THUMBNAILS_DIR / "resampling-nearest-neighbour.png", transparent=True)
```

```python
chart = earthkit.plots.Map(domain=[-18, -10.1, 45, 50], figsize=(2, 2))

chart.pcolormesh(
    rgg,
    resample=Regrid(out_grid={"grid": [0.5, 0.5]}),
    units="celsius",
    colors="Reds",
    levels=np.arange(8.4, 12.4, 0.2),
)
chart.scatter(
    rgg,
    marker="X",
    edgecolors="black",
    units="celsius",
    cmap="Reds",
    levels=np.arange(8.4, 12.4, 0.2),
)
# chart.legend()
chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR / "resampling-regrid.png", transparent=True)
```

```python
chart = earthkit.plots.Map(figsize=(2, 2))

chart.pcolormesh(rgg, resample=Bilinear(), units="celsius", levels=np.arange(8.4, 12.4, 0.2))
# chart.pcolormesh(x=x, y=y, z=values, style=style, resample=Bilinear())
# chart.scatter(x=x, marker="X", y=y, edgecolors="red", z=values, c=values, style=style)
chart.ax.set_frame_on(False)
```

```python
chart = earthkit.plots.Map(domain=[-18, -10.1, 45, 50], figsize=(2, 2))

chart.pcolormesh(t4, units="celsius", levels=np.arange(8.4, 12.4, 0.2), cmap="Reds")
chart.scatter(
    t2,
    marker="X",
    edgecolors="black",
    cmap="Reds",
    units="celsius",
    levels=np.arange(8.4, 12.4, 0.2),
)
# chart.legend()
chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR / "resampling-subsample.png", transparent=True)
```
