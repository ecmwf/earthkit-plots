## Isolines with quickplot

This section covers plotting well-understood meteorological variables with `quickplot` as isolines.

First, let's get some sample temperature and mean sea level pressure data from the ERA5 dataset and visualise it using quickplot.

```python
import earthkit as ek

data = ek.data.from_source("sample", "era5-2t-msl-1985122512.grib")
data.ls()
```

You can extract the pressure field from this data with the `sel()` method.

```python
pressure = data.sel(short_name="msl")
```

Since mean sea level pressure is a very widely used meteorological variable, **earthkit-plots** has a default style available for visualising it. Note that you still need to pass the units you would like to see your data in.

```python
ek.plots.quickplot(pressure, units="hPa").show()
```
