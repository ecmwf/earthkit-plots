## Isolines with contour

This section explores basic plotting with `contour()`.

First, let's get the sample mean sea level pressure data from the ERA5 dataset that we used in the last example.

```python
import earthkit as ek

data = ek.data.from_source("sample", "era5-2t-msl-1985122512.grib")
pressure = data.sel(short_name="msl")
```

Let's plot this data over Europe using the `contour()` method. This will produce a basic representation of the data as contour lines, using the original units of the data (in our case, Pascals).

```python
chart = ek.plots.Map(domain="Europe")
chart.contour(pressure)
chart.land()
chart.coastlines()
chart.title()
chart.show()
```
