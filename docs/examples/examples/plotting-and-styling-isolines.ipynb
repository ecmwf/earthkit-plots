## Plotting and styling isolines

This notebook demonstrates how to use **earthkit** to load and visualise pressure data - or any data that you would like to visualise as isolines, or contour lines of constant value.

We will explore basic plotting with `contour()`, choosing contour levels, and customising styles for better presentation.

First, let's get some sample mean sea level pressure data from the ERA5 dataset and visualise it using quickplot.

```python
import earthkit as ek

temperature, pressure = ek.data.from_source("sample", "era5-2t-msl-1985122512.grib")
pressure.ls()
```

```python
chart = ek.plots.Map(domain="Europe")
chart.contour(pressure, units="hPa", auto_style=True)
chart.legend()
chart.land()
chart.coastlines()
chart.title()
chart.show()
```

### Adding labels, choosing levels and adding highlights

To enhance readability, we can apply contour levels with customised line widths at the levels we choose. We can also label those levels with `labels=True`.

You can pass levels as a simple list or range:

```python
chart = ek.plots.Map(domain="Europe")

chart.contour(pressure, units="hPa", labels=True)

chart.land()
chart.borders()
chart.coastlines()

chart.title()

chart.show()
```

```python
chart = ek.plots.Map(domain="Europe")

chart.contour(
    pressure,
    units="hPa",
    levels={"step": 2, "reference": 1000},
    labels=True,
    linecolors="black",
    linewidths=[0.5, 0.5, 0.5, 1, 0.5]
)

chart.land()
chart.borders()
chart.coastlines()

chart.title()

chart.show()
```

### More styling

```python
chart = ek.plots.Map(domain="North Atlantic")

style = ek.plots.styles.Contour(
    units="hPa",
    levels={"step": 4, "reference": 1000},
    labels=True,
    linewidths=1,
    line_colors="turbo",
)

chart.contour(pressure,
    style=style,
)

chart.legend()

chart.land()
chart.borders()
chart.coastlines()

chart.title()

chart.show()
```
