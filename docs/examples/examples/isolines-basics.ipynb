## Isolines - units, levels and colours

This section demonstrates how to change the units, contour levels and colours.

First, let's get the sample mean sea level pressure data from the ERA5 dataset that we used in the last example.

```python
import earthkit as ek

data = ek.data.from_source("sample", "era5-2t-msl-1985122512.grib")
pressure = data.sel(short_name="msl")
```

We can customise the style of this plot in many ways, including the following:

- **units**: for example, we can pass `units="hPa"` to convert to hectopascals.
- **levels**: you can simply pass levels as a list or range, or alternatively as a dictionary containing arguments to the `earthkit.plots.styles.levels.Levels` class. The valid keys are:
  - `step`: the interval between contour lines;
  - `reference`: make sure the levels include this value;
  - `divergence_point`: centres levels around a value (e.g., 0 for positive/negative data).
- **linecolors**: set the colour(s) of the lines.
- **linewidths**: change the thicknesses of alternating lines.

<div class="alert alert-block alert-info">
<strong>Color vs colour:</strong> You might have noticed some inconsistency between spellings of colour/color in this documentation. This documentation is written in <strong>British English</strong>, in which colo<strong>u</strong>r is spelt with a <strong>u</strong>. However, all colour-related arguments passed to <strong>earthkit-plots</strong> functions use the <strong>American English</strong> spelling <code>color</code>. This is because <strong>earthkit-plots</strong> is built on top of matplotlib, which uses American spellings - and so we do the same in <strong>earthkit</strong> to maximise compatibility.
</div>

```python
chart = ek.plots.Map(domain="Europe")

chart.contour(
    pressure,
    units="hPa",
    #levels=range(980, 1040, 4) - this is just a different way to pass levels!
    levels={"step": 2, "reference": 1000},
    linecolors="red",
    linewidths=[0.5, 1, 0.5, 0.5],
)

chart.land()
chart.borders()
chart.coastlines()

chart.title()

chart.show()
```

<div class="alert alert-block alert-info">
<strong>NOTE: </strong> You can pass any arguments that <a href='https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html#matplotlib-pyplot-contour' target="_blank">matplotlib's contour function</a> accepts here - try experimenting with a few!
</div>
