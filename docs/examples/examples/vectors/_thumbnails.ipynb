```python
import earthkit.data as ekd
import earthkit.plots as ekp
from earthkit.plots.resample import Bilinear, Subsample
from pathlib import Path

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")

data = ekd.from_source("sample", "storm_ophelia_wind_850.grib").to_fieldlist()
data.ls()
```

```python
from earthkit.plots.resample import Bilinear

chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2))
chart.barbs(data, resample=Bilinear(8), color="#F94144", linewidth=0.8)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR/"vectors-basic-barbs.png", transparent=True)
```

```python
from earthkit.plots.resample import Bilinear

chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2))
chart.quiver(data, resample=Bilinear(8), color="#F94144", linewidth=0.8, width=0.01, scale=500)
chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR/"vectors-basic-quiver.png", transparent=True)
```

```python
from earthkit.plots.resample import Bilinear

chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2))
chart.streamplot(data, resample=Bilinear(10), color="#F94144")
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR/"vectors-streamplot.png", transparent=True)
```

```python
from earthkit.plots.resample import Subsample

chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2))
chart.grid_points(data[0], c="cornflowerblue")
chart.quiver(data, resample=Subsample(2, mode="stride"), color="#F94144", linewidth=0.8, width=0.01, scale=500)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR/"vectors-subsampling.png", transparent=True)
```

```python
from earthkit.plots.resample import Subsample

chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2), crs="NorthPolarStereo")
chart.grid_points(data[0], c="cornflowerblue")
chart.quiver(data, resample=Bilinear(8), color="#F94144", linewidth=0.8, width=0.01, scale=500)
chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR/"vectors-regridding.png", transparent=True)
```

```python
from earthkit.plots.resample import Bilinear
style = ekp.styles.Style(
    colors="plasma",
    levels=range(10, 50),
    units="m s-1",
)
chart = ekp.Map(domain=[-18, -10, 45, 50], size=(2, 2))
chart.barbs(data, resample=Bilinear(8), linewidth=0.8, style=style)
chart.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR/"vectors-styles.png", transparent=True)
```
