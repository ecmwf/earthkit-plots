```python
from pathlib import Path

import earthkit.data as ekd

import earthkit.plots as ekp

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")

data = ekd.from_source("sample", "era5-2t-msl-1985122512.grib").to_fieldlist()
pressure = data.sel({"parameter.variable": "msl"})
```

```python
colors = [
    "#FFF0F0",
    "#FFD0CF",
    "#FFA8A6",
    "#FF7B78",
    "#F94144",
    "#E02E31",
    "#C41E21",
    "#A31014",
    "#7E060A",
    "#580006",
]
```

```python
chart = ekp.Map(domain=[-18, -10.1, 45, 50], figsize=(2, 2))
chart.contourf(pressure, units="hPa", levels={"step": 4}, colors=colors)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "contourf-introduction.png", transparent=True)
```

```python
from earthkit.plots.styles import Hatched

HATCHES = ["." * (i + 2) for i in range(8)]

style = Hatched(
    colors=colors,
    hatches=HATCHES,
    legend_style="disjoint",
)

chart = ekp.Map(domain=[-18, -10.1, 45, 50], figsize=(2, 2))
chart.contourf(pressure, units="hPa", style=style)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "contourf-styles.png", transparent=True)
```

```python
chart = ekp.Map(domain=[-18, -10.1, 45, 50], figsize=(2, 2))
chart.contourf(
    pressure,
    units="hPa",
    levels={"step": 4},
    cmap="Spectral",
)
chart.contour(
    pressure,
    labels=False,
    linestyles=["dashed", "solid"],
    colors="#F94144",
    units="hPa",
    levels={"step": 4},
)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "contourf-contouroverlay.png", transparent=True)
```
