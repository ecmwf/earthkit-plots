```python
import earthkit.data as ekd
import earthkit.plots as ekp
from earthkit.geo import regrid
import cartopy.crs as ccrs

from pathlib import Path

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")
```

```python
source = hp = ekd.from_source("sample", "H8_nested_t2.grib2")
```

```python
H8 = regrid(source.to_fieldlist(), grid={"grid": "H8"}, interpolation="linear")
O8 = regrid(source.to_fieldlist(), grid={"grid": "O8"}, interpolation="linear")
LL = regrid(hp.to_fieldlist(), grid={"grid": [10, 10]}, interpolation="linear")
```

```python
chart = ekp.Map(crs=ccrs.NearsidePerspective(central_latitude=51, central_longitude=0), size=(2, 2))
chart.grid_cells(LL, style="auto")

chart.save(THUMBNAILS_DIR/"grid-types-regular-ll.png", transparent=True)
```

```python
chart = ekp.Map(crs=ccrs.NearsidePerspective(central_latitude=51, central_longitude=0), size=(2, 2))
chart.grid_cells(O8, style="auto")
chart.save(THUMBNAILS_DIR/"grid-types-reduced-gg.png", transparent=True)
```

```python
chart = ekp.Map(crs=ccrs.NearsidePerspective(central_latitude=51, central_longitude=0), size=(2, 2))
chart.grid_cells(H8, style="auto")
chart.save(THUMBNAILS_DIR/"grid-types-healpix.png", transparent=True)
```
