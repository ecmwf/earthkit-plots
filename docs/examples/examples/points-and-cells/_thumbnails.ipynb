```python
from pathlib import Path

import earthkit.data as ekd

import earthkit.plots as ekp

THUMBNAILS_DIR = Path("../../../_static/thumbnails/")

data = ekd.from_source("sample", "healpix-h128-nested-2t.grib")
```

```python
colors = [
    "#f94144",
    "#ff5f59",
    "#ff7c72",
    "#ff968a",
    "#ffada2",
    "#ffc2ba",
    "#ffd6d0",
    "#ffebe8",
][::-1]
colors = [
    "#7D0000",
    "#A80A0A",
    "#F94144",
    "#F95F2A",
    "#F97820",
    "#F99214",
    "#E8A800",
    "#D4B800",
    "#C8C200",
    "#C8C800",
][::-1]
```

```python
style = ekp.styles.Style(units="celsius", levels=range(4, 17), colors="Spectral_r")
```

```python
chart = ekp.Map(domain="Spain", figsize=(2, 2))
chart.point_cloud(data, s=10, style=style)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "point-cloud.png", transparent=True)
```

```python
chart = ekp.Map(domain="Spain", figsize=(2, 2))
chart.grid_points(data, s=10, c="#F94144")
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "grid-points.png", transparent=True)
```

```python
chart = ekp.Map(domain="Spain", figsize=(2, 2))
chart.grid_cells(data, style=style)
chart.ax.set_frame_on(False)

chart.save(THUMBNAILS_DIR / "grid-cells.png", transparent=True)
```

```python
obs = ekd.from_source(
    "url",
    "https://get.ecmwf.int/repository/test-data/metview/gallery/madis-maritime.nc",
)

style = ekp.styles.Style(units="celsius", colors="Spectral_r", levels=range(0, 30, 2), extend="both")

my_map = ekp.Map(domain=[-145, -70, 10, 75], figsize=(2, 2))

my_map.scatter(
    obs,
    x="longitude",
    y="latitude",
    z="temperature",
    s=8,
    style=style,
    source_units="K",
)
my_map.land(color="cornflowerblue", alpha=0.2)
my_map.ax.set_frame_on(False)
chart.save(THUMBNAILS_DIR / "scatter.png", transparent=True)
```
