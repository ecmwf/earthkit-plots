`earthkit-plots` provides a number of shortcuts for more quickly applying things like ancillary layers to multiple subplots at once.

```python
import earthkit.data
import earthkit.plots
```

```python
data = earthkit.data.from_source("file", "/Users/mavj/Downloads/efas_dis.nc")
```

```python
data.ls()
```

```python
my_map = earthkit.plots.Map(domain="Europe")

my_map.contourf(data[0], transform_first=True)

my_map.title()

my_map.show()
```
