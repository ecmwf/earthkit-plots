```python
import earthkit.data

import earthkit.plots
```

```python
data = earthkit.data.from_source(
    "url",
    "https://get.ecmwf.int/repository/test-data/earthkit-regrid/test-data/global_0_360/O32.grib",
)
```

```python
LEVELS = [1, 3, 5, 7, 9, 11, 13, 15, 20, 30, 50, 75, 100]
LEVELS = [0 - level for level in LEVELS[::-1]] + [0] + LEVELS

STYLE = earthkit.plots.styles.Style(
    units="celsius",
    levels=range(-40, 41, 4),
    extend="both",
    colors="turbo",
)
```

```python
chart = earthkit.plots.Map(domain="Europe")

chart.contourf(data, style=STYLE)

chart.coastlines()
chart.borders()
chart.gridlines()

chart.legend()

chart.show()
```
