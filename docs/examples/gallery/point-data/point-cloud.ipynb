```python
import earthkit.data
import earthkit.plots
```

```python
data = earthkit.data.from_source("file", "point_cloud.nc")
ds = data.to_xarray()
ds
```

```python
values = ds["10 metre U wind component"].values
x = ds.x.values
y = ds.y.values
```

```python
chart = earthkit.plots.Map()
chart.contourf(z=values, x=x, y=y, colors="viridis")
chart.legend(location="right")
# chart.scatter(x=x, y=y, marker=".", color="#ffffff00")
chart.coastlines()
chart.borders()
chart.gridlines()
chart.show()
```
