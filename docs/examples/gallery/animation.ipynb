```python
from earthkit.plots.components.animations import Animation, Figure
import earthkit.data
from IPython.display import HTML
```

```python
temperature = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-single-levels',
    {
        'product_type': 'reanalysis',
        'variable': '2m_temperature',
        'year': '1985',
        'month': '12',
        'day': '25',
        'time': [f"{hour:02d}" for hour in range(24)],
    },
)
```

```python
temperature.ls()
```

```python
animation = Animation()

animation.add_map()
animation.contourf(temperature, units="celsius")

animation.subplot_titles("ERA5 {variable_name} at {time:%H:%M on %-d %B %Y}")
animation.coastlines()
animation.legend()

HTML(animation.show())
```
