## Ocathedral interpolated

This example demonstrates how plotting [octahedral](https://confluence.ecmwf.int/plugins/viewsource/viewpagesrc.action?pageId=192923347) data with `contourf` will automatically interpolate the grid points into smooth contours.

```python
import earthkit.data as ekd
import earthkit.plots as ekp

data = ekd.from_source(
    "url",
    "https://get.ecmwf.int/repository/test-data/earthkit-regrid/test-data/global_0_360/O32.grib",
)

chart = ekp.Map(domain="Antarctica")

chart.contourf(data, units="celsius", auto_style=True)
chart.grid_points(data)

chart.title()
chart.legend()

chart.coastlines()
chart.gridlines()

chart.show()
```
