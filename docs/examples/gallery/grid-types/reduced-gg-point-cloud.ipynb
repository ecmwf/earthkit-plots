## Ocathedral point cloud

This example demonstrates how plotting [octahedral](https://confluence.ecmwf.int/plugins/viewsource/viewpagesrc.action?pageId=192923347) data with `scatter` will plot the centroid of each grid cell as a point.

```python
import earthkit.data as ekd
import earthkit.plots as ekp

data = ekd.from_source(
    "url",
    "https://get.ecmwf.int/repository/test-data/earthkit-regrid/test-data/global_0_360/O32.grib",
)

chart = ekp.Map(domain="Antarctica")

chart.point_cloud(data, units="celsius", auto_style=True)

chart.title()
chart.legend()

chart.coastlines()
chart.gridlines()

chart.show()
```
