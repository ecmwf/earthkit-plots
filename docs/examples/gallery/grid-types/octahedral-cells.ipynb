## Octahedral four ways

```python
import earthkit.data
import earthkit.plots
```

```python
data = earthkit.data.from_source(
    "url",
    "https://get.ecmwf.int/repository/test-data/earthkit-regrid/test-data/global_0_360/O32.grib",
)

data.ls()
```

```python
fig = earthkit.plots.Figure(rows=2, columns=2, size=(8, 8), domain="Arctic")

for method in ("grid_points", "point_cloud", "grid_cells", "contourf"):
    subplot = fig.add_map()
    getattr(subplot, method)(data, units="celsius", auto_style=True)
    subplot.title(f"${method}$".replace("_", "\_"))

fig.legend(location="right")

fig.coastlines()
fig.gridlines()

fig.show()
```
