## HEALPix point cloud

```python
import earthkit.data
import earthkit.plots
import numpy as np

data = earthkit.data.from_source("sample", "healpix-h128-ring-2t.grib")

chart = earthkit.plots.Map(domain=["France", "Spain"])

chart.point_cloud(
    data,
    levels=np.arange(0, 20, 0.5),
    colors="Spectral_r",
    units="celsius",
)

chart.title()
chart.legend()

chart.coastlines()

chart.show()
```
