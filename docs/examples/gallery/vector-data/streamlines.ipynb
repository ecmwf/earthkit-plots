## Streamlines from wind U and V components

```python
import earthkit as ek
```

```python
data = ek.data.from_source("url", "https://get.ecmwf.int/repository/test-data/metview/gallery/advection_850.grib")
```

```python
data.ls()
```

```python
chart = ek.plots.Map(domain=[-24, 10, 38, 60], crs="PlateCarree")

chart.streamplot(u=data.sel(short_name="u"), v=data.sel(short_name="v"))

chart.land()
chart.coastlines()
chart.gridlines()

chart.title("{variable_name} at {level} hPa - {time:%-d %B %Y %H UTC}")

chart.show()
```
