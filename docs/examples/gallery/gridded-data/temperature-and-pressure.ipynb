```python
import earthkit.data
import earthkit.plots
import cartopy.crs as ccrs

data = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-single-levels',
    {
        'product_type': 'reanalysis',
        'variable': '2m_temperature',
        'year': '1985',
        'month': '12',
        'day': '25',
        'time': '12:00',
    },
)

meridional_mean = data.to_xarray().mean("longitude").squeeze()
zonal_mean = data.to_xarray().mean("latitude").squeeze()
```

```python
temperature_style = earthkit.plots.styles.Style(
#     units="celsius",
#     levels=range(-40, 41, 4),
    extend="both",
    colors="turbo",
)
```

```python
chart = earthkit.plots.Map(domain=["Norway", "Sweden"])

chart.block(data, style=temperature_style)

chart.legend(location="right")

chart.coastlines()

chart.title()
chart.gridlines()


chart.show()
```
