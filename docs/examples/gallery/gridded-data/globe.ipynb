```python
import earthkit.data
import earthkit.plots

import cartopy.crs as ccrs
```

```python
data = earthkit.data.from_source(
    "mars",
    **{
        'channel': '3/2/1',
        'class': 'od',
        'date': -1,
        'ident': '71',
        'instrument': '210',
        'param': '260512',
        'step': '12',
        'stream': 'oper',
        'time': '00:00:00',
        'type': 'ssd',
        'grid': "0.1/0.1",
    },
)
```

```python
chart = earthkit.plots.Map(domain="Arctic")

chart.rgb_composite(data)

chart.show()
```
