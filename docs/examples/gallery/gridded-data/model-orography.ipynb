## Model Orography

```python
import earthkit.data
import earthkit.plots
```

```python
data = earthkit.data.from_source("url", "https://get.ecmwf.int/repository/test-data/metview/gallery/model_orog.grib")
data.ls()
```

```python
levels = [
    0,
    20,
    100,
    200,
    300,
    500,
    800,
    1000,
    1500,
    2000,
    2500,
]

style = earthkit.plots.styles.Style(
    levels=levels,
    extend="both",
    colors="terrain",
    scale_factor=1/9.81,
)
```

```python
chart = earthkit.plots.Map(domain=[-15, 35, 32, 72])

chart.block(data[0], style=style)

chart.ocean(color="#444")
chart.coastlines(color="white")
chart.borders()

chart.gridlines()

chart.legend(label="model orography ($m$)")

chart.show()
```
