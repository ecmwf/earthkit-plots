## Spaghetti plot

```python
import earthkit as ek

ds_en = ek.data.from_source("sample", "ens_storm_st_jude.grib")
```

```python
z_en = ds_en.sel(param="z", level=850, step=78)
```

```python
chart = ek.plots.Map() 

chart.spaghetti(z_en, level=12500, highlight={"dataType": "cf"})
    
chart.land()
chart.coastlines()
chart.borders()
chart.gridlines()

chart.title(
    "ECMWF Run: {base_time:%Y-%m-%d %H} UTC (+{lead_time}h) {variable_name} {level} hPa",
)

chart.show()
```
