```python
import earthkit.plots

chart = earthkit.plots.Map(domain="Japan")

chart.land()
chart.ocean()
chart.urban_areas()
chart.coastlines()
chart.borders()
chart.administrative_areas(resolution="high")
chart.gridlines(color="white", linestyle=":")
chart.cities(adjust_labels=True)

chart.title("Cities in {domain}")

chart.show()
```
