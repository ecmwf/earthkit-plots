```python
import earthkit.plots

chart = earthkit.plots.Map(domain="Arctic")

chart.image("blue-marble.jpg", extent=(-180, 180, -90, 90))

chart.coastlines(color="white", linewidth=0.5)
chart.borders(color="white")

chart.show()
```
