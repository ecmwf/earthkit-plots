## Using a cartopy CRS

This example demonstrates how earthkit-plots can accept any cartopy [Cartopy coordinate system object](https://cartopy.readthedocs.io/stable/reference/projections.html#cartopy-projections) as the `crs` argument to the `Map` class.

```python
import earthkit.plots as ekp
import cartopy.crs as ccrs

my_map = ekp.Map(
    crs=ccrs.NearsidePerspective(
        central_latitude=51,
        central_longitude=10,
    )
)

my_map.stock_img()
my_map.borders()
my_map.coastlines()

my_map.show()
```
