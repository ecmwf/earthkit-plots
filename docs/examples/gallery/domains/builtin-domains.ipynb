## Built-in named domains

This example demonstrates how you can easily create maps from named domains, like country or continent names, in earthkit-plots. See the [Named domains](https://earthkit-plots.readthedocs.io/en/latest/domains-gallery.html) page for a full list of built-in named domains.

```python
import earthkit.plots as ekp

fig = ekp.Figure(rows=2, columns=3)

for domain in [
    "Brazil",
    "North Atlantic",
    "Antarctica",
    "Africa",
    "Greenland",
    "Australia",
]:
    fig.add_map(domain=domain)

fig.land()
fig.borders()
fig.coastlines()
fig.gridlines()

fig.subplot_titles("{domain}")

fig.show()
```
