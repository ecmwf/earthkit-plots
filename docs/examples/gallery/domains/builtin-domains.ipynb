```python
import earthkit.plots

fig = earthkit.plots.Figure(rows=2, columns=3)

for domain in [
    "Brazil",
    "North Atlantic",
    "Antarctica",
    "Africa",
    "Greenland",
    "Australia",
]:
    fig.add_map(domain=domain)

fig.coastlines()
fig.borders()
fig.stock_img()
fig.gridlines()

fig.subplot_titles("{domain}")

fig.show()
```
