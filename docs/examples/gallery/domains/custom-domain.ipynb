```python
import earthkit.plots
from earthkit.plots.geo import domains
```

```python
DOMAIN = domains.Domain.from_bbox(
    bbox=[2, 24, 45, 55],
    name="Central Europe",
)

chart = earthkit.plots.Map(domain=DOMAIN)

chart.land()
chart.borders()
chart.coastlines()
chart.gridlines()

chart.title("{domain}")

chart.show()
```
