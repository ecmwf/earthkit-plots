## Domain with projection

This example demonstrates how to use a named domain with a specific projection/coordiante system, instead of the default coordinate system for that domain. Any [Cartopy projection object](https://cartopy.readthedocs.io/stable/reference/projections.html#cartopy-projections) is compatible with any named domain.

Experiment with the values of the `domain` and `crs` kwargs to see how it affects the plot.

```python
import earthkit.plots as ekp
import cartopy.crs as ccrs

chart = ekp.Map(
    domain="Europe",
    crs=ccrs.NorthPolarStereo(central_longitude=10),
)

chart.coastlines()
chart.borders()
chart.land()
chart.gridlines()

chart.title("{domain} on a {crs} projection")

chart.show()
```
